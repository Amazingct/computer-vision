```python
import cv2 as cv #import openCV package
from matplotlib import pyplot as plt #import pyplot module in matplotlib package
%matplotlib inline
```

```python
#reading images with openCV
img = cv.imread("casava.jpg") # read as BGR
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
```
