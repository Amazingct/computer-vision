```python
#Read as other format
gray = cv.imread('data/glass.jpeg', 0)
```

```python
gray.shape
```

```python
plt.imshow(gray, cmap='gray')
```

```python
#RESIZE
resized = cv.resize(gray,(700,500))
plt.imshow(resized, cmap='gray')
```

```python
#FLIP
flipped = cv.flip(resized,-1)
plt.imshow(flipped, cmap='gray')
```

```python
flipped = cv.flip(resized,0)
plt.imshow(flipped, cmap='gray')
```

```python
flipped = cv.flip(resized,1)
plt.imshow(flipped, cmap='gray')
```

```python
# SAVE IMAGE
cv.imwrite('data/new_img.jpeg', flipped)
```
