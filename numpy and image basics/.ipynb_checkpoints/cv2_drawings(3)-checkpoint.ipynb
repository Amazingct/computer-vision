```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Create a black image
canvas = np.zeros((512,512,3),dtype=np.int16)
```

```python
canvas.shape
```

```python
plt.imshow(canvas)
```

```python
a=cv2.rectangle(canvas,(384,10),(500,150), (0,255,0),5)
```

```python
plt.imshow(canvas)
```

```python
a=cv2.circle(canvas,(100,100),50, (130,255,0),5)
```

```python
plt.imshow(canvas)
```

```python
a=cv2.circle(canvas,(150,100),50, (130,255,0),-1) # -1 thickness fills the image
plt.imshow(canvas)
```

```python
a=cv2.line(canvas,(0,0),(512,512), (130,25,5),5) # -1 thickness fills the image
plt.imshow(canvas)
```
