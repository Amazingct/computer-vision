```python
#template matching
full = cv.imread("dog.jpg")
full = cv.cvtColor(full, cv.COLOR_BGR2RGB)
plt.imshow(full)
```

```python
head = full[0:1000,600:1500,  :]
```

```python
plt.imshow(head)
```

```python
cv.imwrite("head.jpg", head)
```

```python
methods = [cv.TM_CCOEFF, cv.TM_CCOEFF_NORMED, cv.TM_CCORR, cv.TM_CCORR_NORMED, cv.TM_SQDIFF, cv.TM_SQDIFF_NORMED]
```

```python
for m in methods():
    #create a copy of full
    full_copy = full.copy()
    method = eval(m)
    
    res = cv.matchTemplate(full_copy, head, method)
    min_v, max_v, min_loc, max_loc = cv.minMaxLoc(res)
    
    if method in [cv.TM_SQDIFF, CV.TM_SQDIFF_NORMED]:
        top_left = min_loc #(x,y)
    else:
        top_left = max_loc
    
    height, width, channels = head.shape
    bottom_right = (top_left[0]+width, top_left[1]+height)
    
    cv.rectangle(full_copy, top_left, bottom_right, (255,0,0), 10)
    
    # plot
    plt.subplot(121)
    plt.imshow(res)
    plt.title("HEATMAP")
    
    
    plt.subplot(121)
    plt.imshow(full_copy)
    plt.title("DETECTION")
    
    plt.suptitle(m)
    plt.show()
    
    print('\n')
    print('\n')
    
    
    
```
